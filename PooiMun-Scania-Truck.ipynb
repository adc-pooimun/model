## Data Pre-processing

#### Dealing with missing value

```python
null_values = train.isna().sum().div(train.shape[0]).mul(100).to_frame().sort_values(by=0, ascending = False)
null_values.plot.bar(figsize=(50,20))
plt.show()
```

```python
null_values[null_values[0]>75]
```

There are 6 features contain more than 70% missing values. These features will be dropped before training

```python
# Drop features more than 70% missing values
feature_to_drop = list(null_values[null_values[0]>75].index)
train = train.drop(feature_to_drop, axis=1)
test = test.drop(feature_to_drop, axis=1)
```

```python
train.shape
```

```python
test.shape
```

#### Dealing with the remaining missing value with imputation

```python
train.fillna(train.median(), inplace=True)
test.fillna(train.median(), inplace=True)
```

```python
train.isnull().values.any()
```

```python
test.isnull().values.any()
```

```python
# Detect non-numeric features
train.dtypes.values
```

There are no non-numeric features detected

#### Dealing with Imbalance data class with oversampling method

```python
train['class'] = train['class'].replace('neg', value = 0)
train['class'] = train['class'].replace('pos', value = 1)

test['class'] = test['class'].replace('neg', value = 0)
test['class'] = test['class'].replace('pos', value = 1)
```

```python
train['class'].value_counts()
```

```python
test['class'].value_counts()
```

event rate = 1/59 (1.7%)

```python
train_0 = train[train['class']==0]
train_1 = train[train['class']==1]
```

```python
# Resampling class 1
from sklearn.utils import resample
train_1 = resample(train_1,replace=True,n_samples=20000,random_state=1)
```

```python
train = pd.concat([train_0, train_1])
```

```python
train['class'].value_counts()
```

event rate for new dataset =20/59 (34%)

## Data Standardization and train/test splitting

```python
y = train['class']
x = train.drop(['class'],axis=1)
y_test_final = test['class']
X_test_final = test.drop(['class'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x)
train = scaler.transform(x)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_test_final)
train = scaler.transform(X_test_final)
```

```python
# split training testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

## Cost calculation

```python
from sklearn.metrics import make_scorer

def my_scorer(y_test, y_pred):
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    cost = 10*fp+500*fn
    return cost
```

## Logistic Regression

```python
%%time
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

logisticR = LogisticRegression()
logisticR.fit(X_train, y_train)

y_pred = logisticR.predict(X_test)

print('Accuracy on train set: {:.2f}'.format(logisticR.score(X_train, y_train)))
print('Accuracy on test set: {:.2f}'.format(logisticR.score(X_test, y_test)))


y_pred_proba = logisticR.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label= 'AUC = %0.2f' % auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.title('Receiver Operating Characteristic')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
print(' ')
print('Classification Report')
print(classification_report(y_test, y_pred))
print(' ')
print('Confusion Matrix')
print(confusion_matrix(y_test, y_pred))
print(' ')
print('Cost on LogisticR')
print(my_scorer(y_test, y_pred))
```

## Conclusion

Logistic Regression, Random Forest Classifier and Gradient Boosting Classifer have been used to train the model. Result shows that Random Forest Classifier and Gradient Boosting Classifier perform higher accuracy and higher recall. Although Random Forest Classifier give the least cost incurred, this model show overfitting result. 

Higher recall (true positive rate) gives a measure of the model accurately to identify the true positive. In this case study, sensitivity to identify faulty truck is more important to avoid higher cost from truck breakdown

The computation time is longer on Gradient Boosting Classifier but the prediction from Gradient Boosting classifier gives the least cost incurred compared to Logistic regression. 

Principle Component Analysis has been done to estimate the number of component that explained variance of the model. Result shows that around 10 components from the data describe close to 100% of the variance of the model. PCA has been used to transform the data and applied to various models. PCA transformed data does not give better accuracy and does not minimize the cost incurred but it shows a shorter computation time as the features are optimal in PCA transformed data.

We need to strike a balance between computation cost and operational cost. Since the objective of this exercise is to minimize the operational cost, Gradient Boosting Classifier (without PCA tranformed data) will be selected for model development
